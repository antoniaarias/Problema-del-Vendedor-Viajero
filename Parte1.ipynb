```python
from gurobipy import *

# Valor de las variables:
N=5
M=15
Mp=2
Me=2
Ma=3
Tp=4
Te=3
Ta=2
T=3
# Vector de los empleados "I"
I=list(range(1,M+1))
# Vector de las reuniones "J"
J=list(range(1,N+1))
#Modelo:
m=Model("Problema 1.c)")
#Creaci贸n de las variables binarias:
#Xij vale 1 si el trabajador i es presentador en la j reunion
Xij={}
for i in I:
    for j in J:
        Xij[i,j]=m.addVar(vtype=GRB.BINARY)
#Yij vale 1 si el trabajador i es entrevistador en la j reunion
Yij={}
for i in I:
    for j in J:
        Yij[i,j]=m.addVar(vtype=GRB.BINARY)
#Zij vale 1 si el trabajador i es orador en la j reunion
Zij={}
for i in I:
    for j in J:
        Zij[i,j]=m.addVar(vtype=GRB.BINARY)
#Oij vale 1 si el trabajador i es oyente en la j reunion
Oij={}
for i in I:
    for j in J:
        Oij[i,j]=m.addVar(vtype=GRB.BINARY)
#Restricciones:
#1) 1 rol por persona para cada reunion
for j in J:
    for i in I:
        m.addConstr(Xij[i,j]+Yij[i,j]+Zij[i,j]+Oij[i,j]==1) 
#2) todos deben participar en 1 rol al menos 1 vez
m.addConstrs(quicksum(Xij[i,j] for j in J) >=1 for i in I )
m.addConstrs(quicksum(Yij[i,j] for j in J) >=1 for i in I )
m.addConstrs(quicksum(Zij[i,j] for j in J) >=1 for i in I )
#3) cantidad de roles por reunion
m.addConstrs(quicksum(Xij[i,j] for i in I) >=Mp for j in J )
m.addConstrs(quicksum(Yij[i,j] for i in I) >=Me for j in J )
m.addConstrs(quicksum(Zij[i,j] for i in I) >=Ma for j in J )
# Funci贸n objetivo:
m.setObjective(quicksum(quicksum(Xij[i,j]*(T-Tp)+Yij[i,j]*(T-Te)+Zij[i,j]*(T-Ta) for j in J) for i in I), GRB.MAXIMIZE)
m.update()
m.optimize()
print ("Valor optimo=", m.ObjVal)
A=[]
for i in I:
    for j in J:
        #print("Xij=",Xij[i,j].X)
        A=A+[Xij[i,j].X]
print("Xij=",A)
A=[]
for i in I:
    for j in J:
        #print("Yij=",Yij[i,j].X
        A=A+[Yij[i,j].X]
print("Yij=",A)
A=[]
for i in I:
    for j in J: 
        #print("Zij=",Zij[i,j].X)
        A=A+[Zij[i,j].X]
print("Zij=",A)
A=[]
for i in I:
    for j in J:
        #print("Oij=",Oij[i,j].X)
        A=A+[Oij[i,j].X]
print("Oij=",A)
```

```python
#funci贸n que entrega el rol de la persona i en la reuni贸n j
def rol(i,j):
    if Xij[i,j].x==1:
        return "presentador"
    if Yij[i,j].x==1:
        return "entrevistador"
    if Zij[i,j].x==1:
        return "anotador"
    if Oij[i,j].x==1:
        return "oyente"
#funci贸n que entrega la lista con los roles de la persona i en las 5 reuniones
def rolk(i):
    A=[]
    for j in J:
        A=A+[j,"rol que asume="+rol(i,j)]
    return A
#roles de cada persona en cada reuni贸n
for i in I:
    print("Roles que asume la persona:",i)
    print(rolk(i))
```
